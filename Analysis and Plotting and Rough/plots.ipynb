```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
SingleThreadedPerfRef   = [0.011/1000,276.279/1000,5862.26/1000,66889.4/1000,1029360/1000,8252090/1000]
SingleThreadedPerfOpt   = [0.016/1000,371.757/1000,2987.28/1000,24352.5/1000,195646/1000,1547510/1000,(1547510*8)/1000]
MultiThreadedPerfOpt    = [0.211/1000,99.325/1000,793.491/1000,6043.61/1000,48616.3/1000,390769/1000,3192900/1000]
configurations          = [16,512,1024,2048,4096,8192,16384]
```

```python
SingleThreadedPerfRef  = [0.011/1000,276.279/1000,5862.26/1000,66889.4/1000,1029360/1000,8252090/1000,(8252090*8)/1000]
SingleThreadedPerfOpt  = [0.016/1000,371.757/1000,2987.28/1000,24352.5/1000,195646/1000,1547510/1000,(1547510*8)/1000]
MultiThreadedPerfOpt  = [0.211/1000,99.325/1000,793.491/1000,6043.61/1000,48616.3/1000,390769/1000,3192900/1000]
```

```python
configurations      = [16,512,1024,2048,4096,8192,16384]
plt.figure(figsize=(16, 12), dpi=200)
plt.grid()
plt.plot(configurations,SingleThreadedPerfOpt, 'o-g', label = 'Optimised Single Thread')
plt.plot(configurations[:-1],SingleThreadedPerfRef[:-1], 'o-r',label = 'Naive Single Thread')
plt.plot(configurations,MultiThreadedPerfOpt, 'o-b',label = 'Optimised Multithreaded')
plt.legend()
plt.ylabel('Time in Seconds.(s)')
plt.xlabel('Dimension of matrices.(N)')
plt.title('Trend of execution time for RMM')
```

```python
configurations      = [16,512,1024,2048,4096,8192,16384]
plt.figure(figsize=(16, 12), dpi=200)
plt.grid()
plt.plot(configurations,np.log(np.array(SingleThreadedPerfOpt)), 'o-g', label = 'Optimised Single Thread')
plt.plot(configurations[:-1],np.log(np.array(SingleThreadedPerfRef[:-1])), 'o-r',label = 'Naive Single Thread')
plt.plot(configurations,np.log(np.array(MultiThreadedPerfOpt)), 'o-b',label = 'Optimised Multithreaded')
plt.legend()
plt.ylabel('Time in Seconds.(in log_10 scale)')
plt.xlabel('Dimension of matrices.(N)')
plt.title('Trend of execution time for RMM')
```

```python
configurations      = [16,512,1024,2048,4096,8192,16384]
plt.figure(figsize=(16, 12), dpi=200)
plt.grid()
plt.loglog(configurations,SingleThreadedPerfOpt, 'o-g', label = 'Optimised Single Thread')
plt.loglog(configurations[:-1],SingleThreadedPerfRef[:-1], 'o-r',label = 'Naive Single Thread')
plt.loglog(configurations,MultiThreadedPerfOpt, 'o-b',label = 'Optimised Multithreaded')
plt.legend()
plt.ylabel('Time in Seconds.(in log_10 scale)')
plt.xlabel('Dimension of matrices.(N) in log10 scale')
plt.title('Trend of execution time for RMM')
```
